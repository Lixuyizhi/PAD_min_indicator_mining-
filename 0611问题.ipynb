### 1. 数据质量问题
- 分钟级文本数据量不足，导致情绪信号稀疏

- 评论数据质量参差不齐，噪音较大
- 不同品类的评论数据量差异大，可能导致偏差

### 2. 方法论问题
- 情绪衰减模型可能过于简化
- 开盘时刻的情绪释放机制需要更细致的建模
- 信号量计算方法可能需要更复杂的权重体系
- PAD三维度的组合方式可能需要非线性映射

### 3. 市场微观结构影响
- 分钟级别的价格波动受到市场微观结构的强烈影响
- 做市商行为、流动性因素可能掩盖了情绪因子的效果
- 高频交易的存在可能干扰了情绪-价格的传导机制

### 4. 时间尺度不匹配
- 情绪形成和传导可能需要更长的时间周期
- 分钟级别可能过于短期，无法捕捉情绪的完整影响
- 不同市场状态下情绪传导速度可能不同

### 5. 理论基础的局限
- 情绪对资产价格的影响机制可能比预想的更复杂

- 需要考虑更多的市场环境因素和条件概率

### 6. 技术实现的挑战
- 数据处理和对齐过程中可能引入了偏差
